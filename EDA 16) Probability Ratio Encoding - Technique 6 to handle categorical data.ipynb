# Probability Ratio Encoding
Steps
1) Calculate probability of "Survived" wrt "Cabin" which is a categorical feature

2) Calculate probability of "Died" wrt "Cabin" which is a categorical feature

3) Calculate probability of "Survived" / prob"Died"

4) Create a Dictionary to map cabin with probability

5) Replace with the categorical features / values

```python
import pandas as pd

df = pd.read_csv('titanic_train.csv')
df
```

```python
import pandas as pd

df = pd.read_csv('titanic_train.csv',usecols=['Cabin', 'Survived'])
df
```

```python
# Replacing Nan values with "Missing" in the column "Cabin"
df['Cabin'].fillna('Missing',inplace=True)
df
```

```python
df['Cabin'].unique() # List of Unique categories in the column / feature "Cabin"
```

```python
# extracting the first letter
# to create blocks e.g - C,B,E , etc

df['Cabin'] = df['Cabin'].astype(str).str[0]
df
```

```python
#Method 1
df['Cabin'].unique() # List of Unique categories in the column / feature "Cabin"
```

```python
#Method 2
df.Cabin.unique() # List of Unique categories in the column / feature "Cabin"
```

```python
# defniing probability dataframe
prob_df = df.groupby(['Cabin'])['Survived'].mean() # calculating the "Survived" probability of person in each "cabin"/block
prob_df
```

```python
prob_df = pd.DataFrame(prob_df) #converting the data in Dataframe
prob_df #changed the values in the column "Survived"
```

```python
# Calculating the probability that the person has died
prob_df['Died'] = 1 - prob_df['Survived']
prob_df
```

# Probability Ratio = prob(Survived) / prob(died)

```python
# Calculating the probability ratio
prob_df['Probability_Ratio'] = prob_df['Survived'] / prob_df['Died']
prob_df
```

```python
# Converting the probability ratio into Dictionary
probability_encoded = prob_df['Probability_Ratio'].to_dict()
probability_encoded
```

```python
# creating New column to map "probability_encoded" with "Cabin"
df['Cabin_encoded'] = df['Cabin'].map(probability_encoded)
df
```

# Advantage
1) Easy to imlement

2) "Cabin_encoded" in numerical value is more useful than "Cabin" in categorical values
